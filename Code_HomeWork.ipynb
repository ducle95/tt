```python
def caculate_factorial(number):
    result =1;
    for i in range (1,number+1):
        result*=i
    return result

def estimate_e_number():
    result =0
    n =0
    while(n<=100):
        n+=1
        result +=1/ caculate_factorial(n)
    return 1+result

estimate_e_number()
```

```python
import math
def caculate_factorial(number):
    result =1;
    for i in range (1,number+1):
        result*=i
    return result
def caculater_sinx(x,n):
    result =0.0
    for i in range(0,n):
        result+=pow(-1,i)*pow(x,2*i+1)/caculate_factorial(2*i+1)
    return result

def caculater_cosx(x,n):
    result =0.0
    for i in range(0,n):
        result+=pow(-1,i)*pow(x,2*i)/caculate_factorial(2*i)
    return result

def caculater_sinh(x,n):
    result =0.0
    for i in range(0,n):
        result+=pow(x,2*i+1)/caculate_factorial(2*i+1)
    return result

def caculater_coshx(x,n):
    result =0.0
    for i in range(0,n):
        result+=pow(x,2*i)/caculate_factorial(2*i)
    return result

caculater_cosx(math.pi/3,20)
```

```python
def reverse_number(number):
    my_list = []
    while(number !=0):
        temp_number = number%10
        my_list.append(str(temp_number))
        number//=10
    if(my_list[0] == 0):
        my_list = my_list[1:]
        
        
    result = ''.join(my_list)
    return int(result)

number = 1253070
reverse_number(number)   
```

```python
import math
def cal_sigmoid(x):
    result =0
    result = 1/(1+ math.exp(-x))
    return result

def cal_tanh(x):
    result =0
    result = (math.exp(x)- math.exp(-x))/(math.exp(x)+ math.exp(-x))
    return result
cal_sigmoid(1)
cal_tanh(2)
```
